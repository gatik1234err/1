```python


plt.figure(figsize=(10,10))
plt.style.use('dark_background')

# Create crazy spiral pattern
theta = np.linspace(0, 50*np.pi, 5000)
r = np.linspace(0.1, 2, 5000)

x = r * np.cos(theta)
y = r * np.sin(theta)

colors = np.sin(theta*0.3) + np.cos(theta*0.2)

plt.scatter(x, y, c=colors, cmap="plasma", s=5, alpha=0.9)

plt.title("Psychedelic Spiral Chaos", fontsize=20, weight='bold')
plt.axis("off")

plt.show()
```
